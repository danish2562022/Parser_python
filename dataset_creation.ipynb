# Dataset Creation

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("cleaned_data.csv")
```

```python
data=data.drop(['Unnamed: 0'],axis=1)
data.head()
```

```python
price=data['price-btcUSD']
price_1_days_after=price[1:]
price_7_days_after=price[8:]
price_28_days_after=price[29:]
```

```python

data['price_1_days_after']=price_1_days_after.tolist()+[0]*1
data['price_7_days_after']=price_7_days_after.tolist()+[0]*8
data['price_28_days_after']=price_28_days_after.tolist()+[0]*29
```

```python
data.tail(30)
```

```python
data.drop(data.tail(29).index,
        inplace = True)
data.tail()
```

```python
x=data['price_1_days_after']>data['price-btcUSD']
data['Price_icrease_or_decrease_after_1_day']=x.astype(int)
x=data['price_7_days_after']>data['price-btcUSD']
data['Price_icrease_or_decrease_after_7_day']=x.astype(int)
x=data['price_28_days_after']>data['price-btcUSD']
data['Price_icrease_or_decrease_after_28_day']=x.astype(int)
```

```python
data.to_csv("Final_data.csv",index=False)
```
